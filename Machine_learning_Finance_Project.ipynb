---
**BASICS**

*   K-means clustering is a popular unsupervised machine learning algorithm used for partitioning a dataset into k clusters based on similarity measures. It aims to minimize the sum of squared distances between each data point and its assigned centroid (the mean of all points in the cluster).
*   Classification：K-means clustering can be classified as a partitional clustering technique, which means that it divides a dataset into non-overlapping subsets, where each data point belongs to only one cluster. It is also a centroid-based clustering method because it uses the mean (centroid) of the data points within each cluster to represent that cluster.

---
**DISADVANTAGES**

*   Sensitivity to initial centroids: K-means clustering is sensitive to the initial placement of centroids. The algorithm may converge to a suboptimal solution if the initial centroids are not chosen carefully. 

*   Determination of the optimal number of clusters: The number of clusters (k) needs to be determined before applying the algorithm, which can be a challenging task in practice. If the wrong number of clusters is chosen, the resulting clustering may not provide useful insights.

*   Dependence on data distribution: K-means clustering assumes that the clusters are spherical and equally sized, which may not always be the case in real-world data. It may not work well with clusters that have irregular shapes or different densities.

*   Outlier sensitivity: K-means clustering is sensitive to outliers, which can significantly affect the placement of centroids and, consequently, the resulting clustering.

*   Difficulty handling categorical data: K-means clustering is designed for continuous variables and may not be suitable for categorical variables, as it relies on calculating means and distances.

---
**FEATURES**

*   Handles missing values: K-means clustering can handle missing values by assigning them to the nearest centroid.

*   Scalable: K-means clustering is scalable and can handle large datasets with many variables and observations.

*   Speed: K-means clustering is computationally efficient and can converge quickly to a solution.

*   Requires a priori knowledge of the number of clusters: The number of clusters (k) needs to be determined before applying the algorithm, which can be a challenging task in practice. If the wrong number of clusters is chosen, the resulting clustering may not provide useful insights.

*   Convergence: The algorithm outputs whether or not it has converged to a stable solution, based on a predetermined convergence criterion, such as the change in SSE between iterations.

**EQUATIONS**

Mathematically, the objective of the K-means algorithm is to minimize an objective function which in our case is a squared error function given by 
$$J=\sum_{j=1}^k \sum_{i=1}^n ||x_i^{j} - C_j||^2$$
where 

$$||x_i^{j} - C_j||^2$$
is the Euclidean distance between data points $x_i$ and the centroid $C_j$.

Below is the mathematical walkthrough of the algorithm:
*   Suppose we have a set of data points X={x1,x2,...,xn} and a set of centroids V={v1,v2,...,vc}. The first step in the K-Means algorithm is to randomly select centroids from the data set. Next, we compute the distance between each centroid and the data points. Based on this distance, we assign each data point to the nearest centroid. We then recalculate the centroids by taking the mean of the data points assigned to each group. This process of assigning data points to the closest centroid and recalculating the centroids is repeated until convergence.
$$v_i = \left(\frac{1}{c_i}\right)\sum_{j=1}^{c_i} x_j$$
where $c_i$ is the number of data points in the ith cluster.
*   Find the distance between the new centroid and each data point.
*   If there is no change in the composition of the clusters then we stop, or else we repeat the procedure.
One limitation of the K-means algorithm is that it requires the number of clusters to be specified in advance. To address this issue, we can use methods such as the Silhouette score or the elbow method to estimate the optimal number of clusters. These techniques can help us select the number of clusters that best balances the trade-off between minimizing the within-cluster distance and maximizing the between-cluster distance.

```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# Download and prepare data
data = yf.download(["AAPL", "IBM"], start="2021-01-01", end="2022-12-31")['Close']
X = data.pct_change().dropna().values


# Visualize the dataset
X_blobs, y_blobs = make_blobs(centers=4, n_samples=3000)  # create blobs for visualization only
fig = plt.figure(figsize=(8, 6))
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=y_blobs)
plt.title("Dataset with 4 clusters", fontsize=20)
plt.xlabel("First feature")
plt.ylabel("Second feature")
plt.suptitle(
    "Fig. 5: Dataset Visualization", fontweight="bold", horizontalalignment="right"
)
plt.show()
```

```python
# Perform K-Means clustering with elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(
        n_clusters=i, init="k-means++", max_iter=300, n_init=10, random_state=42
    )
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot elbow method to determine optimal number of clusters
plt.plot(range(1, 11), wcss)
plt.title("K-Means Clustering (The Elbow Method)", fontsize=20)
plt.xlabel("K")
plt.ylabel("WCSS")
plt.suptitle("Fig. 6: The Elbow Method", fontweight="bold", horizontalalignment="right")
plt.show()
```

```python
# Perform K-Means clustering with selected number of clusters
num_clusters = 4  # choose the number of clusters based on the elbow method
kmeans = KMeans(n_clusters=num_clusters, init="k-means++", max_iter=300, n_init=10, random_state=0)
y_means = kmeans.fit_predict(X)

# Visualize the clustering results
plt.scatter(X[y_means == 0, 0], X[y_means == 0, 1], s=100, c="red", label="Cluster 1")
plt.scatter(X[y_means == 1, 0], X[y_means == 1, 1], s=100, c="blue", label="Cluster 2")
plt.scatter(X[y_means == 2, 0], X[y_means == 2, 1], s=100, c="green", label="Cluster 3")
plt.scatter(X[y_means == 3, 0], X[y_means == 3, 1], s=100, c="orange", label="Cluster 4")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c="yellow", label="Centroids")
plt.title("Cluster of Dataset")
plt.xlabel("AAPL returns")
plt.ylabel("IBM returns")
plt.legend()
plt.show()
```

## Advantages
PCA is an unsupervised learning method that has many advantages. For instance, PCA helps us improve the generalizability of machine learning models by lowering the complexity of the data. This aids in our ability to overcome the "curse of dimensionality".

Performance of algorithms is frequently influenced by the size of the data. High-dimensional data models may operate very slowly or even crash, consuming a lot of server resources. At a negligible loss of model accuracy, PCA can assist us in enhancing efficiency.

The ability to generate independent, uncorrelated features of the data and the reduction of noise in the data are additional advantages of PCA. Additionally, PCA enables data visualization and the examination of clustering/classification methods. 

# Basics

PCA is a machine learning algorithm for dimensionality reduction. Principal component analysis, also known as PCA, is a statistical technique that transforms a collection of observations of potentially correlated variables into principal components, a set of values of linearly uncorrelated variables.

Formally, PCA is a statistical method for lowering a dataset's complexity. To achieve this, the data are transformed linearly into a new coordinate system, where (most) of the variation in the data can be described with fewer dimensions than the original data.

# Disadvantages

Firstly, principal components are poorly interpreted. Principal components are linear combinations of the initial data's features, but they are more difficult to understand. For instance, it can be challenging to determine the dataset's most crucial characteristics after computing principal components. 

Secondly, there is a trade-off between knowledge loss and dimensionality reduction. Dimensionality reduction has a price even though it is helpful. Loss of information is an essential component of PCA. Unfortunately, when using PCA, we must choose between dimensionality reduction and information loss, and this trade-off must be balanced.

# Equations

The linear transformation of the features can be represented by the equation $$y = WX$$ where $W$ are the weights, $X$ the original features and $y$, the principal components.

In matrix form, finding PCA can be expressed as $$Y= XA$$

The rows of the matrix $A$ are the eigenvectors and they describe how the principal components are oriented compared to the original features. The composition of the eigenvectors are the weights $a_{ij}$ and are called the loadings. Loadings give us information on how each feature contributes to a principal component.

Loadings range from $-1$ to $1$ and large loadings (close to $1$ or $-1$) indicate the presence of a strong relationship between a particular feature and the principal component whereas loadings closer to zero indicate a weaker relationship. The sign of the loading shows whether a feature is positively or negatively correlated to the principal component.

Given a matrix $A$ and the variance-covariance matrix of our original data, $S_X$, the variance-covariance matrix of the resulting principal components will be expressed as $$S_Y = A S_X A^T$$

# Features

PCA makes the assumption that features are correlated. PCA won't be able to identify the main components if the features (or dimensions, or columns, in the case of tabular data) are not correlated.

PCA is not resistant to outliers. The algorithm will be biased in datasets with significant outliers. For this reason, it is advised to eliminate outliers before running PCA.

A linear relationship between features is assumed by PCA. Non-linear connections cannot be captured effectively using the algorithm. It is therefore recommended to convert non-linear features or links between features into linear features using common techniques like log transforms.

Technical implementations of PCA frequently presumptively assume that all values are present. It's common practice to presume that the feature set has no missing values when computing PCA using statistical software tools (i.e. no empty rows). An approximation should be used to impute missing values or remove the rows and/or columns with empty values (e.g. the mean of the column).

The output of PCA are the eigenvectors and eigenvalues that define the orientation and length of the ellipsoid that best fits the dataset.

# Step 3 (Technical Section)
## PCA
The total explained variance is obtained by adding the explained variance ratios of the first N components. Choosing the number of dimensions for which the cumulative explained variance surpasses a cutoff point, such as 0.95 (95%) or 0.8 (80%), is a typical course of action for PCA.

In the literature, numerous criteria have been put forth in an effort to determine the ideal amount of dimensions for PCA. Jolliffe (2002) describes three categories of methods: Ad-hoc principles (like the scree test), tests based on distributional hypotheses, and computational techniques like bootstrapping, permutation, and cross-validation.

# Step 4 (Marketing Alpha)
## PCA
Advantages of PCA include:
- Simple to calculate. The foundation of PCA is linear algebra, which computers can handle well computationally.
- Other machine learning methods are accelerated through usage of PCA. When trained on principal components rather than the initial dataset, machine learning algorithms converge more quickly.
- Reduces the problems caused by high-dimensional data. Regression-based algorithms are readily overfit when dealing with high-dimensional data. We can avoid the predictive algorithms overfitting by using PCA in advance to reduce the dimensionality of the training dataset. 
- Promotes better visualization. Understanding and visualizing data in high dimensions is very challenging. PCA can reduce high-dimensional data to low-dimensional data (e.g. two dimensions), allowing for simple visualization.

# Disadvantages

- It can be computationally expensive and time-consuming to train on large datasets with many features.
- The resulting models can be difficult to interpret, as it can be challenging to determine which features are most important.
- Random Forest may not perform as well when there are strong correlations between features, as this can result in overfitting and reduced predictive accuracy.
- It may also struggle to make accurate predictions when the data is imbalanced, as it tends to prioritize the majority class over the minority class.

# Guide

**Inputs:**

- Training dataset: a dataset with both input features and output labels used to train the model.
- Hyperparameters: parameters that can be set before training the model, such as the number of trees, maximum depth of the trees, and the number of features to consider when splitting nodes.
- Test dataset: a dataset with input features but no output labels used to evaluate the model's performance.

**Outputs:**

- Trained model: a model that has learned the relationship between the input features and output labels in the training dataset.
- Predictions: predictions of the output label for each input feature in the test dataset, generated by applying the trained model to the test dataset.

# Hyperparameters

- **n_estimators:** The number of trees in the forest.
- **max_features:** The maximum number of features considered for splitting a node. The default value is sqrt(n_features) for classification tasks and log2(n_features) for regression tasks.
- **max_depth:** The maximum depth of each tree in the forest. If None, then nodes are expanded until all the leaves are pure or until all the leaves contain less than min_samples_split samples.
- **min_samples_split:** The minimum number of samples required to split an internal node. The default value is 2.
- **min_samples_leaf:** The minimum number of samples required to be at a leaf node. The default value is 1.
- **criterion:** The function to measure the quality of a split. The default value is "gini" for classification tasks and "mse" for regression tasks.

**Other less commonly tuned hyperparameters include:**

- **bootstrap:** Whether or not to use bootstrap samples when building trees. The default value is True.
- **oob_score:** Whether or not to use out-of-bag samples to estimate the generalization accuracy. The default value is False.
- **class_weight:** The weight associated with each class. The default value is None.
- **random_state:** The seed used by the random number generator. The default value is None.
